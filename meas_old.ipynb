```python
def read_data(path):
    file = open(path, 'r', encoding='ISO-8859-14')
    x = []
    y = []
    for i, line in enumerate(file):
        if i>103:
            p = line.split()
            x.append(float(p[0]))
            y.append(float(p[1]))
    file.close()
    return x, y
```

```python
def gate_voltage(path):
    file = open(path, 'r', encoding='ISO-8859-14')
    for i, line in enumerate(file):
        if i==38:
            voltage = float(line[7:-2])
    file.close()
    return voltage
```

```python
def plot_procedure(root, files):
    dacs = []
    numbers = []
    paths = []
    for file in files:
        filename = os.path.join(root, file)
        paths.append(filename)
        numbers.append(transistor_number(filename))
        dacs.append(abscissa_type(filename))
    for i in range (1,9):
        for index, no in enumerate(numbers):
            if numbers[index]==i and dacs[index]=='DAC 1':
                x,y = read_data(paths[index])
                gate = gate_voltage(paths[index])
                number = transistor_number(paths[index])
                measurment_type = abscissa_type(paths[index])
                
                ax = plt.gca()
                plt.scatter(x, y, c='r', label = measurment_type+' '+str(number))
                plt.xlabel('$U_{sd}, V$')
                plt.ylabel('$I_{sd}, A$')
                plt.title(paths[index]+'\t'+str(number))
                ax.text(max(x)/3, min(y)*2, f'$V_g = {int(gate/1000)} V$',
                        horizontalalignment='left',
                        verticalalignment='bottom',
                        transform=ax.transAxes)
                plt.legend()
                plt.show()
```

```python
for (root,dirs,files) in os.walk(PATH, topdown=True):
    plot_procedure(root, files)
```
